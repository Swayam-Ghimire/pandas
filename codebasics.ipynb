# fillna, interpolate

```python
import pandas as pd
import numpy as np
from datetime import datetime
df=pd.DataFrame({
    'Column1': [1, 2, np.nan, 4, 5],
    'Column2': ['A', np.nan, 'C', 'D', 'E'],
    'Column3': [10.5, 20.2, np.nan, 12.1, np.nan],
    'Column4': [np.nan, False, False, np.nan, True],
    'Column5': [datetime(2023, 1, 1), datetime(2023, 1, 2), datetime(2023, 1, 3), datetime(2023, 1, 8), datetime(2023, 1, 9)]
})
```

```python
df
```

```python
df.set_index('Column5', inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
type(df.Column4[1])
```

```python
newdf=df.fillna(0)
newdf
```

```python
newdf=df.fillna({"Column1":3,"Column2":'B',"Column3":2.2,"Column4":True})
newdf
```

```python
newdf=df.fillna(method='ffill')
newdf
```

```python
newdf=df.fillna(method='bfill')
newdf
```

```python
df.interpolate()
```

```python
df.interpolate(method='time')
```

```python
dt=pd.date_range("2023-01-01","2023-01-09")
index=pd.DatetimeIndex(dt)
df.reindex(index)
```
