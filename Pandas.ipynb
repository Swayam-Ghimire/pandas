```python
df = pd.DataFrame({"name":['Sommy','Swayam','Subin'],
                 "toy":[np.nan,'Batmobile','Bullwhip'],
                   "born":[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]})
""""np.nan refers to a special floating-point value in the NumPy library, which stands for "Not a Number." It is used to 
represent missing or undefined data in numerical arrays. The nan value is typically used in situations where a particular
value is not available or cannot be represented, such as when there are data gaps, invalid calculations, or incomplete 
measurements."""
```

```python
"""
1.For the "name" column: Since the "name" column is of type object (string), using np.nan or pd.NaT would both represent 
missing values. In this case, there won't be any significant difference between them, and you can choose to use either one.

2.For the "toy" column: Since the "toy" column contains string values, using np.nan will work fine to represent missing values. 
However, if you intend to perform any numerical operations on this column (e.g., calculating statistics or applying mathematical functions), np.nan might cause data type conflicts or errors.

3.For the "born" column: This column contains datetime values. Here, it is recommended to use pd.NaT to represent missing dates 
instead of np.nan. pd.NaT is specifically designed to handle missing or undefined datetime values in pandas and provides 
appropriate functionalities for working with datetime data. Using np.nan in a datetime column may result in unexpected behavior
or errors when performing datetime operations."""
```
